```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sbn

df_men= pd.read_csv("data9b_m.txt", sep="	")
df_women= pd.read_csv("data9b_w.txt", sep="	")

df_men.info()
```

```python
result=df_men.head(5)

print(result)

result2=df_women.head(5)

print(result2)
```

```python
print("The average number of steps taken by men  =",df_men["steps"].mean()) #the average number of steps taken by men 

print("The average number of steps taken by women=",df_women["steps"].mean()) #the average number of steps taken by women 
```

```python
df_men.plot(x="steps", y="bmi", kind="scatter")
df_women.plot(x="steps", y="bmi", kind="scatter")
```

```python
plt.scatter(df_men["steps"], df_men["bmi"], color="red", label="men" )
plt.scatter(df_women["steps"], df_women["bmi"], color="blue", label="women")
plt.xlabel("steps")
plt.ylabel("bmi")
plt.legend()
plt.show()
```

```python
df_women.corr(method ='kendall') #To find the correlation for women dataset the columns using kendall method 
```
